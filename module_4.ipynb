# Practical Tips
It's hard to know in advance where to focus, but it's counter productive to hypothesis and just start building system and experiment with it. 

### Evals
To implement evals:
1. Build a system and localize the error
2. Eval to measure output format: manually extract format to create a pattern and then match it to the output in our prompt
3. Monitor changes in our workflow to see how metric improves

Good tone is to develop a custom code piece to evaluate the outcome of LLM (like max length of the output).


As we discussed before - evals can be **objective** (code) and **subjective** (LLM-as-judge).


### Error analysis
Rather then going by guts - we can examine traces to understand each step in the workflow and localize the error. From % of each module we can conclude which component can be more problematic.

* Audit of traces - analysis to localize the component

For that we need to have 25+ samples and create a comparison to find deviation from normality. 

### Improving component performance
We can enhance non-LLM components (web search, text retrieval RAG, etc) by:
* Tuning h-params
* Replacing the component 

For LLM components:
* Better prompt
* New model
* Split complext instructions into smaller steps
* Model fine-tune (expensive and complex)

Usually playing with models is the best way.

### Latency and cost optimization
First - build the solution, then we need to scale and optimize. There are tools for optimizations:

* Run time - overall vs component track - consider parallelism
* LLM steps - cost per token
* API/compute - cost for tools

### Process summary
Build <-> Analyze | Start building system end-to-end and then build evals and traces to compute metrics.
