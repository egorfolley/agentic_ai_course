# Module 1. Intro to Agentic AI Workflows
* Non-agentic workflow (zero-shot) - a simple prompt as input with simple output
* Agentic workflow - breaking a complex task into a list of commands for thinking/acting and revision of the work that has been done

### Degrees of autonomy
* Simple agents with low autonomy - predetermined steps - command->LLM->actions->output
* More autonomous - makes decisions itself - command->select tools->use tools->LLM->more tools->LLMs for each action->output

Range from less autonomous (rigid) to semi-autonomous (HITL) to highly autonomous (take actions themselves).

### Benefits
They can work autonomously to collect and process more efficiently than human. While autonomous take more time, they give higher benchmark metrics. 
Working in modular systems (different tools) allows flexibility of choice of work.

### Difficulties
* Easy - clear workflows with standard procedures to follow (text input)
* Hard - dynamic environments with planning on the way (multimodal input)

### Task decomposition
To make the most efficiency, we must go deeper into creative planning before developing the system. Breaking down the task is crucial for higher accuracy. 

Building blocks are usually:
* Models 
* Tools for actions 

To build effective agents - think of clear sequence of building blocks to combine them. Clearly separate what can be done by model vs tool.

### Evals
Effective agentic AI depends heavily on the way we evaluating their performance. To build a better evaluation - look for the output and think how to improve each piece of it. Evals helps us to track the error:
* LLM-as-a-judge - subjective evals
* Using code - objective evals

2 types:
* End-to-end 
* Component-level

Tracing LLMs to perform error analysis. 

### Agentic Design Patterns
1. Reflection
- Request and then guide/feedback agent to make improvements

2. Tool use
- Functions use to get the work done 

3. Planning
- Request input and agent decides the steps-to-take

4. Multi-agent collaboration
- Let a team of agents work together on tasks to get higher results
