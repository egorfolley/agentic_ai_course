# Module 2. Reflection Design Patterns
Human reviews and adjust based on missing information. Agent must do the same - LLM to write, then LLM to reflect and improve. 

Reflection is powerful when it has additional information to use for feedback.

### Zero/few-shot prompt
* Zero - no examples
* One/few - # of examples 

Reflection outperforms direct generation. Examles include: missing characters, email assessment, facts/dates, etc. The best way to modify the prompt is to explore other people prompts.

### Reflection evals
In order to maximize the reflection we can evaluate it with the ground truth answer. This way we can measure the output quality. 

* LLM as a judge - subjective tasks
* Grading with a rubric for **consistent results** - this way we grade each criteria to measure the eval 

1. Objective evals (deterministic rules) - code-based are easier + dataset of groudtruth samples - use for correctness
2. Subjective evals (diverse/judgement) - LLM as a judge + rubric-based grading is better - hard to formalize qualities

### Reflection design pattern
Prompt with **no reflection** plateaus, while reflection has higher performance. Reflection **with external feedback** gives the best performance.
