# Module 3. Tool Use
Extra functionality for LLMs.


### Creating a tool
These are just code that can be executed and give the access to the **system prompt**. 

### Tools syntax
Behind the scenes the tools consist of JSON schema with type=function, name, parameters, description. They are helpful for LLMs to understand it's purpose, output, parameters for accurate decision when to use.

### Code execution
We can also add a code execution within the prompt to solve the query


It can be non-secure method (AI can act unpredictably, like deleting files), so sandboxes (Docker) can protect against such errors. Moreover, these days we have MCP protocol.

### MCP 
**Model Context Protocol** that allows LLMs to communicate with external tools, applications, and functions. It solves the fragmentation problem where each tool integration previously required custom, one-off logic.
