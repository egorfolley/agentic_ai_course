# Patterns for Highly Autonomous Agents
Planning workflows are the essential part of system prompt - it makes the strategy more effective. The clarity of each LLM shows the efficiency of the output of this component. 

* Planning with code improves performance (code as action) 

### Multi-agentic workflows
Each task can be assigned to a uniquely-qualified agent. Team can perform better than individuals. 

* **Linear planning** - when we have a chain of agnets
* **Coordinator** - when a major LLM organizing the work of agents

Communication patterns can be challenging, so it's crucial to organize the collaboration between agents:
* Linear - a->a->a
* Hierarchy
* Deeper hierarchy
* All-to-all
